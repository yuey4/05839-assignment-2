```python
import pandas as pd
from collections import defaultdict
from typing import Dict

race_df = pd.read_csv('race_distribution.csv')
for [i] in race_df[['State']].values:
    assert i in state_to_code, '{} not in map'.format(i)
race_distribution: Dict[str, Dict[str, float]] = defaultdict(dict)
columns = race_df.columns.values
for row in race_df.values:
    state, vals = row[0], row[1:]
    for k, v in zip(columns[1:], row[1:]):
        race_distribution[state][k] = v
```
